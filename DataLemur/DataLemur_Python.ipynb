## Convert an Integer to Base 13

**Given an integer `num`, return its string representation in base 13.**

In case you don’t use base 13 that much (who does, right?), here’s a quick rundown: just like base 10 uses digits from `0` to `9`, base 13 uses `0` to `9` **and** for `10`, `11` and `12`, we use the letters `A`, `B`, and `C`.

### For example:

- `9` in base 13 is `"9"`
- `10` in base 13 is `"A"`
- `11` in base 13 is `"B"`
- `12` in base 13 is `"C"`
- `13` in base 13 is `"10"`
- `14` in base 13 is `"11"`
- `49` in base 13 is `"3A"` (since `3 × 13 + 10 = 49`)
- `69` in base 13 is `"54"` (since `5 × 13 + 4 = 69`)


```python
from math import log
def convertToBase13(num):
  '''
  watch this video for best explaination
  https://www.youtube.com/watch?v=s3mxIcr7fOQ
  '''
  sign = 'pos' if num >= 0 else 'neg'
  num = abs(num)
  dic = {i:'0123456789ABC'[i] for i in range(13)}
  if num != 0:
    power = int(log(num,13))
  else:
    return '0'
  result = ''
  for i in range(power,-1,-1):
    result += dic[num//(13**i)]
    num %= (13**i)
  return result if sign == 'pos' else '-' + result
```

## Same Stripe Diagonal Matrix Check

You are given an `m x n` matrix. Your task is to determine if the matrix has **diagonal stripes**, where all elements in each diagonal from **top-left to bottom-right** are identical.

In this context, each diagonal stripe runs from the **top-left corner** to the **bottom-right corner** of the matrix.  
Check if **every diagonal stripe** consists entirely of the same number.

Return `True` if **all diagonals** contain the same value within themselves, otherwise return `False`.

---

### Example 1

**Input:**  
`matrix = [[42, 7, 13, 99], [6, 42, 7, 13], [1, 6, 42, 7]]`

**Output:**  
`True`

**Explanation:**  
In this grid, the diagonals are:
- `[1]`
- `[6, 6]`
- `[42, 42, 42]`
- `[7, 7, 7]`
- `[13, 13]`
- `[99]`

All elements in each diagonal are identical.  
Thus, the answer is `True`.

---

### Example 2

**Input:**  
`matrix = [[8, 23], [69, 1]]`

**Output:**  
`False`

**Explanation:**  
The diagonal `[8, 1]` does not consist of elements of the same stripe.
Therefore, the answer is `False`.


```python
def is_same_stripes(matrix):

  if len(matrix[0]) != len(matrix):
    idx = [[j for j in range(i,-1,-1)] for i in range(len(matrix[0])-1)]

  else:
    idx = [[j for j in range(i,-1,-1)] for i in range(len(matrix[0]))]

  rows = len(matrix) - 1
  diag = []
  for i in idx:
    for j in i:
      diag.append(matrix[rows][j])
      rows = rows - 1
    rows = len(matrix) - 1

    if len(set(diag)) == 1:
      diag = []
      continue
    else:
      return False

  matrix = [i[::-1] for i in matrix][::-1]
  for i in idx:
    for j in i:
      diag.append(matrix[rows][j])
      rows = rows - 1
    rows = len(matrix) - 1

    if len(set(diag)) == 1:
      diag = []
      continue
    else:
      return False

  return True
```

## Weakest Strong Link

You know that phrase, *a chain is only as strong as its weakest link*?  
Imagine you have a **chain-link fence**, represented as a matrix.

For the chain-link at position `(i, j)`, the input matrix `strength[i][j]` indicates how strong the chain is at that position (where **stronger means a higher number**).  
All the numbers in the matrix are **unique**.

---

The **Weakest Strong Link** is defined as:
- The **weakest chain-link in its row**, **and**
- The **strongest link in its column**.

Given a matrix `strength`, **return the Weakest Strong Link** if it exists; otherwise, return `-1`.  
If a Weakest Strong Link exists, there is always **exactly one**, and it can be proven that no other link satisfies both conditions simultaneously.

---

### Example 1

**Input:**  
`strength = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]`

**Output:**  
`7`

**Explanation:**  
`7` is the **weakest** in its row `[7, 8, 9]` and the **strongest** in its column `[1, 4, 7]`, making it the **Weakest Strong Link**.

---

### Example 2

**Input:**  
`strength = [[9, 8, 10], [6, 15, 4]]`

**Output:**  
`-1`

**Explanation:**  
No chain-link satisfies the criteria of being **weakest in its row** and **strongest in its column**.


```python
def weakest_strong_link(strength):
	rows =  [min(i) for i in strength]
	cols = []
	l = []
	for i in range(len(strength[0])):
	  for j in range(len(strength)):
	     l.append(strength[j][i])
	  cols.append(max(l))
	  l.clear()
	try:
	  return list(set(rows).intersection(set(cols)))[0]
	except:
	  return -1
```

  Each contribution increases your balance, which in turn increases the amount of interest you earn next year.

- **Years to Invest**: How long your money stays in the account, compounding annually.

**Your task:**  
Using these four values, calculate the **final amount** you’ll have after the specified number of years with **annual compounding**.

The **triangular sum** of `nums` is the value of the only element left in `nums` after the following process:

1. Let `nums` contain `n` elements. If `n == 1`, end the process.
2. Otherwise, create a new integer array `newNums` of length `n − 1`.
3. For each index `i`, set `newNums[i]` as `(nums[i] + nums[i+1]) % 10`.
4. Replace `nums` with `newNums`.
5. Repeat the process starting from step 1.

Return the **triangular sum** of `nums`.

---

### Test Case 1

**Input:**  
`nums = [1, 3, 5, 7]`

**Iteration 1:**  
`newNums = [(1 + 3) % 10, (3 + 5) % 10, (5 + 7) % 10] = [4, 8, 2]`

**Iteration 2:**  
`newNums = [(4 + 8) % 10, (8 + 2) % 10] = [2, 0]`

**Iteration 3:**  
`newNums = [(2 + 0) % 10] = [2]`

**Output:**

---

### Test Case 2

**Input:**  
`nums = [9, 7, 5, 3]`

**Iteration 1:**  
`newNums = [(9 + 7) % 10, (7 + 5) % 10, (5 + 3) % 10] = [6, 2, 8]`

**Iteration 2:**  
`newNums = [(6 + 2) % 10, (2 + 8) % 10] = [8, 0]`

## Roman Numeral to Integer

Given a valid Roman numeral, convert it to an **integer**.

---

In case you don’t think about the Roman Empire that often, here’s a quick refresher on how to go from a Superbowl `XLIV` to an actual number.

| Symbol | Value |
|--------|-------|
| I      | 1     |
| V      | 5     |
| X      | 10    |
| L      | 50    |
| C      | 100   |
| D      | 500   |
| M      | 1000  |

---

**How it works:**  
Roman numerals are usually written from **largest to smallest**, left to right.  
For example:
- `XI` → `X = 10` and `I = 1` → `10 + 1 = 11`
- `XXX` → `X + X + X = 30` (unless we’re talking about your browser’s search history 👀)

---

**Subtraction rule:**  
In certain cases, a smaller numeral comes **before** a larger one to indicate **subtraction**.

Examples:
- `IV` → `I` before `V` → `5 − 1 = 4`
- `XC` → `X` before `C` → `100 − 10 = 90`

This subtraction rule applies in these cases:
- `I` before `V` (5) or `X` (10) → makes 4 and 9
- `X` before `L` (50) or `C` (100) → makes 40 and 90
- `C` before `D` (500) or `M` (1000) → makes 400 and 900

---

### Example 1

**Input:**  
`s = "XI"`

**Output:**  
`11`

**Explanation:**  
`X = 10`  
`I = 1`  
`XI = 11`

---

### Example 2

**Input:**  
`s = "LXIX"`

**Output:**  
`69`

**Explanation:**  
`L = 50`  
`X = 10`  
`IX = 9`  
`LXIX = 69`. Nice.


```python
def romanToInt(s):
  dic = {'I' : 1, 'V' : 5,'X':10,'L':50,'C':100,'D':500,'M':1000}
  sum = 0
  flag = False
  # CDXX
  for i in range(len(s)):
    if flag == True:
      flag = False
      continue
    if i != len(s) - 1:
      if (s[i+1] ==  'V' or s[i+1] == 'X') and s[i] == 'I':
	      sum += dic[s[i+1]] - 1
	      flag = True
      elif (s[i+1] ==  'L' or s[i+1] == 'C') and s[i] == 'X':
	      sum += dic[s[i+1]] - 10
	      flag = True
      elif (s[i+1] ==  'D' or s[i+1] == 'M') and s[i] == 'C':
	      sum += dic[s[i+1]] - 100
	      flag = True
      else:
	      sum += dic[s[i]]
	      flag = False
    else:
	     sum += dic[s[i]]
  return sum
```

  (e.g., `"one hundred and fifteen"`).

Given an integer `N`, **return the total number of letters used**  
when writing the numbers from `1` to `N` in words.

---

### Example 1

**Explanation:**  
The numbers between `1` and `5` written in words are:
- `1 → "one"` (3 letters)
- `2 → "two"` (3 letters)
- `3 → "three"` (5 letters)
- `4 → "four"` (4 letters)
- `5 → "five"` (4 letters)

Total: `3 + 3 + 5 + 4 + 4 = 19` letters.


```python
def total_letters(N):
	dic = {0:0,1:3, 2:3, 3:5, 4:4, 5:4, 6:3, 7:5, 8:5, 9:4, 10:3, 11:6, 12:6, 13:8, 14:8, 15:7, 16:7, 17:9, 18:8, 19:8, 20:6, 30:6, 40:5, 50:5, 60:5, 70:7, 80:6, 90:6}
	sum = 0
	for i in range(1,N+1):
	  if len(str(i)) == 1:
	    sum += dic[i]
	  elif len(str(i)) == 2:
	    if i >= 10 and i < 20:
	      sum += dic[i]
	    else:
	      sum += dic[(i//10)*10] + dic[(i%10)]
	  elif len(str(i)) == 3:
	   sum += 7
	   sum += dic[i//100]
	   i = i%100
	   sum = sum+3 if i != 0 else sum
	   if i > 10 and i < 20:
	      sum += dic[i]
	   else:
	    sum += dic[(i//10)*10] + dic[(i%10)]
	return sum
```

```python
def is_anagram(s, t):
  if (len(list(s)) == len(list(t))) and (set(list(s)) == set(list(t))):

  	dic_s = {}
  	for i in list(s):
  	  if i in dic_s:
  	    dic_s[i] += 1
  	  else:
  	    dic_s[i] = 1

  	dic_t = {}
  	for i in list(t):
  	  if i in dic_t:
  	    dic_t[i] += 1
  	  else:
  	    dic_t[i] = 1

  	for i,j in dic_s.items():
  	  if j == dic_t[i]:
  	    continue
  	  else:
  	    return False
  	return True

  else:
    return False
```

## Pascal's Triangle

Given an integer `numRows`, return the first `numRows` of **Pascal's Triangle**.

---

In Pascal's Triangle, **each number is the sum of the two numbers directly above it**.

---

**Examples**

- For `numRows = 1`:  
  `[[1]]`

- For `numRows = 4`:  
  `[[1], [1, 1], [1, 2, 1], [1, 3, 3, 1]]`


```python
def generate(numRows):
  l = []
  for i in range(0,numRows):
    if i == 0:
      l.append([1])
    elif i == 1:
	    l.append([1,1])
    else:
      inner = l[i-1]
      num = [1] + [inner[i] + inner[i+1] for i in range(len(inner[:-1]))] + [1]
      l.append(num)
  return l
```

## Longest Consecutive Sequence

Given an array of integers nums, return the length of the longest consecutive sequence of elements that can be formed from the array.

A consecutive sequence consists of elements where each element is exactly 1 greater than the previous element. The elements in the sequence can be selected from any position in the array and do not need to appear in their original order.

Example #1
Input: nums = [100, 4, 200, 1, 3, 2]

Output: 4

Explanation: The longest consecutive is [1, 2, 3, 4] which have a length of 4

Example #2
Input: nums = [3, 2]

Output: 2

Explanation: The longest consecutive is [2, 3] which have a length of 2

```python
def longest_consecutive(nums):
  nums = sorted(set(nums))
  longest_consecutive = 0
  counter = 1
  for i in range(len(nums)):
    try:
      if nums[i]+1 == nums[i+1]:
        counter += 1
      else:
        if counter > longest_consecutive:
          longest_consecutive = counter
        counter = 1
    except:
        if counter > longest_consecutive:
  	      longest_consecutive = counter

  return longest_consecutive
```

It is repeatedly replaced by the sum of the squares of its digits.

The number becomes 1, in which case it is not a looping number.

The number starts repeating in a cycle that does not include 1, making it a looping number.

Return True if n is a looping number, otherwise return False.

Example #1
Input: n=4
Output: True

Explanation:

    4**2=16

    1**2+4**2+5**2=42

```python
def is_looping(n):
  seen = set()

  while True:
    if n in seen:
      return True
    elif n == 1:
      return False
    else:
      seen.add(n)
      n = sum([int(i)**2 for i in str(n)])
```

## Maximize Satisfied Teammates (FAANG Gift Cards)

Imagine you're a VP at FAANG, and you want to reward your team with gifts during Christmas.  
You check with HR and they can only provide a **limited set of gift cards** — each with different values.

---

Each teammate has different **expectations** for their gift:
- Each expects a **minimum gift value**.
- You can only give each person **at most one gift card**.
- If you don’t have a card that meets someone’s minimum expectation, it’s better not to give them one at all.

---

**Goal:**  
Assign each teammate a gift card that meets or exceeds their expectation  
and **maximize** the total number of teammates who are satisfied.

Return the **maximum number** of satisfied teammates.

---

### Example 1

**Input:**  
`Expectations = [5, 20, 1000]`  
`Cards = [10, 15, 100]`

**Output:**  
`2`

**Explanation:**  
- Team member expecting `$5` → get `$10` gift card  
- Team member expecting `$20` → get `$100` gift card  
- Team member expecting `$1000` → no card is big enough

---

### Example 2

**Input:**  
`Expectations = [10, 30, 100, 200]`  
`Cards = [5, 2, 5, 3, 9]`

**Output:**  
`0`

**Explanation:**  
- Even the highest gift card `$9` is less than the lowest expectation `$10`.
- No one is satisfied.


```python
def max_satisfaction(expectations, cards):
  expectations = sorted(expectations)
  cards = sorted(cards)
  count = 0
  for i in expectations:
    idx = []
    for j in cards:
      if j >= i:
        idx.append(cards.index(j))
    if len(idx) > 0:
      cards.pop(idx[0])
      count += 1
    else:
      break
  return count
```

Given two `n x n` **binary matrices** `mat` and `target`,  
**return `True` if it is possible** to make `mat` equal to `target` by rotating `mat` in **90-degree increments**,  
or `False` otherwise.

---

### Example 1

**Input:**  
`mat = [[1,0,1],[0,0,1],[1,0,1]]`  
`target = [[1,0,1],[1,0,0],[1,0,1]]`

**Output:**  
`True`

**Explanation:**

---

### Example 2

**Input:**  
`mat = [[0, 1],[1, 0]]`  
`target = [[1, 1],[0, 0]]`

**Output:**  
`False`

**Explanation:**  
It is impossible to make `mat` equal to `target` by any rotation.


```python
def find_rotation(mat, target):
	l = [[] for i in mat]

	for i in range(len(mat)):
	  for j in range(len(mat)):
	    l[j].append(mat[i][j])

	if l == target:
	  return True

	mat = [i[::-1] for i in mat]

	if mat == target:
	  return True

	l = [[] for i in mat]
	for i in range(len(mat)):
	  for j in range(len(mat)):
	    l[j].append(mat[i][j])

	if l == target:
	  return True
	return False
```

## Spiral Matrix Order

Given an `m x n` matrix, return **all elements** of the matrix in **spiral order**.

---

**What does spiral order mean?**

Start at the **top-left** corner:  
1. Move **right** across the first row.  
2. Move **down** the last column.  
3. Move **left** along the bottom row.  
4. Move **up** the first column.  
Repeat this process inward layer by layer until all elements are collected.

---

It’s like peeling an onion… but way less tear-inducing!

---

### Example 1

**Input:**  
`matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]`

**Output:**  
`[1, 2, 3, 6, 9, 8, 7, 4, 5]`

---

### Example 2

**Input:**  
`matrix = [[1, 2, 3], [8, 9, 4], [7, 6, 5]]`

**Output:**  
`[1, 2, 3, 4, 5, 6, 7, 8, 9]`


```python
import numpy as np
def spiral_order(matrix):
  cnd = len(matrix)+1
  l = []
  try:
    for i in range(len(matrix)+2):
      if i != cnd:
        l.extend(matrix[0])
        matrix = matrix[1:]
        matrix = np.flip(matrix,axis=1).T
      else:
        l.extend(matrix.ravel())
  except:
    pass
  return [int(i) for i in l]
```

- Titles can be shared by multiple people.

---

**Task:**

---

### Example 1

**Input:**  
`answers = [2, 1, 1]`

**Output:**  
`5`

**Explanation:**  
- First respondent: 2 other people ⇒ group size must be 3.
- Second & third respondents: 1 other person ⇒ they likely share the same title ⇒ group size must be 2.
- Minimum total attendees = `3 + 2 = 5`.

---

### Example 2

**Input:**  
`answers = [0, 0, 1]`

**Output:**  
`4`

**Explanation:**  
- First two respondents: 0 ⇒ they each have a **unique** title ⇒ 2 people.
- Third respondent: 1 ⇒ there must be 2 people with this title.
- Minimum total attendees = `2 + 2 = 4`.


```python
def min_attendees(answers):
  dic = {}
  for i in answers:
    if i in dic:
      dic[i] += 1
    else:
      dic[i] = 1
  counter = 0

  for i,j in dic.items():
    if i == 0:
      counter += j
    elif j == 1:
      counter += i+1
    elif j-1 == i:
      counter += i+1
    else:
      grp = j-i
      for _ in range(grp):
        counter += i+1

  return counter
```

```python
import numpy as np

def corr(x, y):
	x = np.array(x)
	y = np.array(y)
	std_x = x.std()
	std_y = y.std()
	cov = np.sum((x-x.mean())*(y-y.mean()))/len(x)-1
	return cov/(std_x*std_y)
```
